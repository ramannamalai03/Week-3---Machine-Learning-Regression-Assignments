```python
import lightgbm
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int)
dataset
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
from lightgbm import LGBMRegressor
regr = LGBMRegressor(boosting_type='gbdt', num_leaves=31, max_depth=-1, learning_rate=0.1, n_estimators=100, subsample_for_bin=200000, 
                      objective=None, class_weight=None, min_split_gain=0.0, min_child_weight=0.001, min_child_samples=20, 
                      subsample=1.0, subsample_freq=0,force_row_wise="true",
                      colsample_bytree=1.0, reg_alpha=0.0, reg_lambda=0.0, random_state=None, n_jobs=None, importance_type='split')
regr.fit(X_train, y_train)
```

```python
y_pred=regr.predict(X_test)
```

```python
from sklearn.metrics import r2_score  
r_score=r2_score(y_test,y_pred)
```
